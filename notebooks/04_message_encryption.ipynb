## Message Encryption Using the BB84 Key

This notebook connects the BB84 key generation simulation to a simple message encryption demo.

The process is:

1. Alice and Bob run BB84.
2. The system calculates QBER.
3. If QBER is above the safety threshold, communication is aborted.
4. If the generated keys match, Alice encrypts the message using her key.
5. Bob decrypts the ciphertext using his key.

For this version, the project uses XOR encryption as an educational one-time-pad style demonstration. This is not production cryptography. Future versions can replace this with AES or another standard encryption method.

This version also does not implement error correction or privacy amplification. If Alice and Bob's keys do not match exactly, the communication is aborted.
