# BB84 Basics Without Eavesdropping

This notebook tests the first version of the BB84 protocol simulator without Eve.

In this version:

- Alice generates random bits.
- Alice generates random bases.
- Bob generates random bases.
- Bob measures Alice's transmitted bits.
- Alice and Bob keep only the positions where their bases match.
- QBER is calculated between Alice and Bob's sifted keys.

When there is no eavesdropper or channel noise, the QBER should be 0.

## Observation

With no eavesdropper and no channel noise, Alice and Bob's sifted keys match exactly.

The sifted key length is usually around half of the original number of qubits because Alice and Bob choose matching bases about half the time.

The QBER should be 0.
