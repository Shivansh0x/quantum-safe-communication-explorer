# Eve Intercept-Resend Attack Simulation

This notebook tests BB84 when Eve performs an intercept-resend attack.

Eve intercepts a selected fraction of qubits. For each intercepted qubit, she randomly chooses a basis, measures the qubit, and resends her measured result to Bob.

Because Eve does not always choose the correct basis, her measurement can disturb the transmitted state. This disturbance appears as errors in Alice and Bob's sifted keys.

## Observation

As Eve intercepts more qubits, QBER increases.

For a simple intercept-resend attack, the expected relationship is approximately:

```text
QBER ≈ Eve interception probability × 25%
```

For example, if Eve intercepts 25% of the qubits, the expected QBER is around 6.25%.
